```python
train_data = pd.read_csv('/kaggle/input/titanic/train.csv')

train_data.head()
```

```python
test_data = pd.read_csv('/kaggle/input/titanic/test.csv')

test_data.head()
```

```python
train_data.isnull().sum()*100 / len(train_data)
```

```python
mean_age = round(train_data['Age'].mean(), 1)
print(mean_age)
```

```python
train_data['Age'].replace(np.nan, mean_age, inplace = True)

train_data.isnull().sum()*100 / len(train_data)
```

```python
mode_embarked = train_data['Embarked'].mode()[0]
print(mode_embarked)
```

```python
train_data['Embarked'].replace(np.nan, mode_embarked, inplace = True)

train_data.isnull().sum()*100 / len(train_data)
```

```python
train_data.drop(columns = 'Cabin', inplace = True)

train_data.head()
```

```python
train_data.drop(columns = ['Name', 'Ticket', 'Embarked'], inplace = True)

train_data.head()
```

```python
train_data = pd.get_dummies(train_data, columns=['Sex'])
train_data.drop('Sex_female', axis = 1, inplace = True)

train_data.head()
```

```python
test_data.isnull().sum()
```

```python
mean_age_test = test_data['Age'].mean()
print(mean_age_test)

test_data['Age'].replace(np.nan, mean_age_test, inplace = True)
test_data.isnull().sum()
```

```python
mean_fare_test = test_data['Fare'].mean()
print(mean_age_test)

test_data['Fare'].replace(np.nan, mean_fare_test, inplace = True)
test_data.isnull().sum()
```

```python
test_data = pd.get_dummies(test_data, columns=['Sex'])
test_data.drop('Sex_female', axis = 1, inplace = True)
```

```python
train_data.head()
```

```python
from statsmodels.formula.api import ols

lr_model = ols(formula = 'Survived ~ Pclass + Sex_male + Age + SibSp + Parch + Fare', data = train_data).fit()
```

```python
lr_model.summary()
# Parch and Fare p-value are higher than 0.05 (alpha value), so we'll remove them
```

```python
train_data.corr()
```

```python
mlr_model = ols(formula = 'Survived ~ Pclass + Sex_male + Age + SibSp', data = train_data).fit()

mlr_model.summary()
```

```python
from sklearn.linear_model import *

prediction = mlr_model.predict(test_data)
```

```python
test_data['Predicted'] = prediction * 100

test_data.head(10)
```
